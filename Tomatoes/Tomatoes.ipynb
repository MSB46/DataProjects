```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

links = []
driver = webdriver.Chrome() 

for page in range(1,3):
    driver.get('https://editorial.rottentomatoes.com/guide/essential-animated-movies/'+str(page)+'/')
    time.sleep(1)
    movie_we = driver.find_elements(By.CLASS_NAME, "article_movie_poster")
    
    
    for i in range(len(movie_we)):
        links.append(movie_we[i].get_attribute('href'))
        
links
```

```python
dta = pd.DataFrame(columns=['Movie Name', 'Year', 'Audience Score', 'Tomatometer Score', 'Rating', 'Genre',
                            'Original Language', 'Director', 'Producer', 'Writer', 'Release Date (Theaters)',
                           'Release Date (Streaming)', 'Box Office (Gross USA)', 'Runtime', 'Distributor'])
```

```python
import random as r

from tqdm import tqdm
for link in tqdm(links, desc='loading....'):
    driver.get(link)
    info_we = driver.find_elements(By.XPATH,'//score-board')
    movie_info = {'Movie Name':'','Year':'','Audience Score':'','Tomatometer Score':''}
    movie_info['Movie Name'] = info_we[0].find_element(By.CLASS_NAME, 'scoreboard__title').text
    movie_info['Year'] = info_we[0].find_element(By.CLASS_NAME, 'scoreboard__info').text.split(',')[0]
    movie_info['Audience Score'] = info_we[0].get_attribute('audiencescore')
    movie_info['Tomatometer Score'] = info_we[0].get_attribute('tomatometerscore')
    webelement_list_val = []
    webelement_list_key = []
    webelement_list_key = driver.find_elements(By.XPATH,'//div[@data-qa="movie-info-item-label"]')
    webelement_list_val = driver.find_elements(By.XPATH,'//div[@data-qa="movie-info-item-value"]')
    key_list, val_list = [],[]
    
    for k,v in zip(webelement_list_key, webelement_list_val):
        key_list.append(k.text.strip(':'))
        val_list.append(v.text)
        
    info = dict(zip(key_list,val_list))
    total_info = {**movie_info,**info} 
    dta = dta.append(total_info, ignore_index=True)



# for _ in range(4):
#     driver.get(links[r.randint(1,139)])
    
#     info_we = driver.find_elements(By.XPATH,'//score-board')

#     movie_info = {'Movie Name':'','Year':'','Audience Score':'','Tomatometer Score':''}
#     movie_info['Movie Name'] = info_we[0].find_element(By.CLASS_NAME, 'scoreboard__title').text
#     movie_info['Year'] = info_we[0].find_element(By.CLASS_NAME, 'scoreboard__info').text.split(',')[0]
#     movie_info['Audience Score'] = info_we[0].get_attribute('audiencescore')
#     movie_info['Tomatometer Score'] = info_we[0].get_attribute('tomatometerscore')
    
#     webelement_list_key = driver.find_elements(By.XPATH,from tqdm import tqdm
for link in tqdm(links, desc='loading....'):
    driver.get(link)
    info_we = driver.find_elements(By.XPATH,'//score-board')
    movie_info = {'Movie Name':'','Audience Score':'','Tomatometer Score':''}
    movie_info['Movie Name'] = info_we[0].text.split('n')[0]
    movie_info['Audience Score'] = info_we[0].get_attribute('audiencescore')
    movie_info['Tomatometer Score'] = info_we[0].get_attribute('tomatometerscore')
    webelement_list_val = []
    webelement_list_key = []
    webelement_list_key = driver.find_elements(By.XPATH,'//div[@data-qa="movie-info-item-label"]')
    webelement_list_val = driver.find_elements(By.XPATH,'//div[@data-qa="movie-info-item-value"]')
    key_list, val_list = [],[]
    for k,v in zip(webelement_list_key, webelement_list_val):
        key_list.append(k.text.strip(':'))
        val_list.append(v.text)
    info = dict(zip(key_list,val_list))
    total_info = {**movie_info,**info} 
    dta = dta.append(total_info, ignore_index=True)from tqdm import tqdm
for link in tqdm(links, desc='loading....'):
    driver.get(link)
    info_we = driver.find_elements(By.XPATH,'//score-board')
    movie_info = {'Movie Name':'','Audience Score':'','Tomatometer Score':''}
    movie_info['Movie Name'] = info_we[0].text.split('n')[0]
    movie_info['Audience Score'] = info_we[0].get_attribute('audiencescore')
    movie_info['Tomatometer Score'] = info_we[0].get_attribute('tomatometerscore')
    webelement_list_val = []
    webelement_list_key = []
    webelement_list_key = driver.find_elements(By.XPATH,'//div[@data-qa="movie-info-item-label"]')
    webelement_list_val = driver.find_elements(By.XPATH,'//div[@data-qa="movie-info-item-value"]')
    key_list, val_list = [],[]
    for k,v in zip(webelement_list_key, webelement_list_val):
        key_list.append(k.text.strip(':'))
        val_list.append(v.text)
    info = dict(zip(key_list,val_list))
    total_info = {**movie_info,**info} 
    dta = dta.append(total_info, ignore_index=True)'//div[@data-qa="movie-info-item-label"]')
#     webelement_list_val = driver.find_elements(By.XPATH,'//div[@data-qa="movie-info-item-value"]')

#     for k,v in zip(webelement_list_key, webelement_list_val):
#         key_list.append(k.text.strip(':'))
#         val_list.append(v.text)

#     info = dict(zip(key_list,val_list))
    
#     total_info = {**movie_info,**info}
#     dta = dta.append(total_info, ignore_index=True)
```
