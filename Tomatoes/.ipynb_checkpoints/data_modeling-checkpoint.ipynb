```python
import pandas as pd
import re
import warnings
import numpy as np
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', 100)

df = pd.read_csv("imdb_eda.csv")
df.head()
```

```python
df.columns
```

# Column Selection

```python
df_model = df.reindex(columns=['worldwide_gross_usd','avg_rating_score', 'votescore',
       'metacritic', 'rating', 'year', 'runtime_minutes', 'genre_count', 'story_word_count', 'prod_company_count','aspect_ratio', 'country_count',
       'language_count', 'votes','director_count', 'writer_count', 'genre_action', 'genre_adventure', 'genre_biography', 'genre_comedy',
       'genre_crime', 'genre_drama', 'genre_family', 'genre_fantasy',
       'genre_history', 'genre_horror', 'genre_music', 'genre_musical',
       'genre_mystery', 'genre_romance', 'genre_sci-fi', 'genre_sport',
       'genre_thriller', 'genre_war', 'genre_western', 'desc_girl', 'desc_boy',
       'desc_young', 'desc_family', 'desc_friend', 'desc_save', 'desc_evil',
       'company_disney', 'company_dreamworks', 'company_pixar',
       'company_columbia', 'company_dentsu', 'company_sony',
       'company_universal', 'company_fox', 'company_ghibli',
       'company_paramount', 'company_other', 'continent_namerica',
       'continent_europe', 'continent_asia', 'continent_oceania',
       'lang_english', 'lang_japanese', 'lang_french', 'lang_spanish',
       'lang_german'])
```

```python
df_model
```

```python
# Worldwide Box Office
df_model_a = df_model[df_model.worldwide_gross_usd > 0]

# Votescore Rating
df_model_b = df_model
```

# Dummy Data (A)

```python
df_dummy_a = pd.get_dummies(df_model_a)
```

```python
df_dummy_a
```

# Test Train Split (A)

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

X = df_dummy_a.drop(['worldwide_gross_usd'],axis=1)
y = df_dummy_a.worldwide_gross_usd.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)

X_sm = sm.add_constant(X)
model = sm.OLS(y, X_sm).fit()

print(model.summary())
```

# Multiple Linear Regression (A)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn import metrics

lm_a = LinearRegression()
lm_a.fit(X_train,y_train)

# sorted(metrics.SCORERS.keys())
np.mean(cross_val_score(lm_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=3))
```

# Lasso (A)

```python
from sklearn.linear_model import Lasso
la_alpha = []
la_results = []
for x in range(1, 100):
    la_alpha.append(x/100)
    la_a = Lasso(alpha=x/100)
    la_a.fit(X_train, y_train)
    
    la_results.append(np.mean(cross_val_score(la_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=3)))
    
print(f'alpha = {(la_results.index(max(la_results)) + 1) * .10} \n{max(la_results)}')

plt.plot(la_alpha,la_results)
```

# Random Forest (A)

```python
from sklearn.ensemble import RandomForestRegressor

rf_a = RandomForestRegressor(random_state=10)
rf_a.fit(X_train, y_train)

np.mean(cross_val_score(rf_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=3))
```

# Grid Search CV (A)

```python
from sklearn.model_selection import GridSearchCV

parameters = {'n_estimators':range(10,100,10), 'max_depth':range(3,10), 
              'criterion':('mse','mae'), 
              'max_features':(None, 'sqrt', 'log2')}

gs_a = GridSearchCV(rf_a, parameters, scoring='neg_mean_absolute_error', cv=3)
gs_a.fit(X_train,y_train)
```

```python
gs_a.best_score_
```

```python
gs_a.best_estimator_
```

# Ensemble Testing (A)

```python
pred_lm = lm_a.predict(X_test)
pred_la = la_a.predict(X_test)
pred_gs = gs_a.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, pred_lm))
print(mean_absolute_error(y_test, pred_la))
print(mean_absolute_error(y_test, pred_gs))

print(mean_absolute_error(y_test, (pred_gs + pred_la)/2))
```
