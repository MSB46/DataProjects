# Target Encoding and Test Train Split (A)

```python
# TargetEncoder classes written by Anurag Bantu:
# https://www.kaggle.com/code/anuragbantu/target-encoding-beginner-s-guide


class KFoldTargetEncoderTrain(base.BaseEstimator, base.TransformerMixin):
    def __init__(
        self, colnames, targetName, n_fold=5, verbosity=False, discardOriginal_col=False
    ):
        self.colnames = colnames
        self.targetName = targetName
        self.n_fold = n_fold
        self.verbosity = verbosity
        self.discardOriginal_col = discardOriginal_col

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        mean_of_target = X[self.targetName].mean()
        kf = KFold(n_splits=self.n_fold, shuffle=False)
        col_mean_name = self.colnames
        X[col_mean_name] = np.nan
        for tr_ind, val_ind in kf.split(X):
            X_tr, X_val = X.iloc[tr_ind], X.iloc[val_ind]
            X.loc[X.index[val_ind], col_mean_name] = X_val[self.colnames].map(
                X_tr.groupby(self.colnames)[self.targetName].mean()
            )
            X[col_mean_name].fillna(mean_of_target, inplace=True)
        if self.verbosity:
            encoded_feature = X[col_mean_name].values
            print(
                "Correlation between the new feature, {} and, {} is {}.".format(
                    col_mean_name,
                    self.targetName,
                    np.corrcoef(X[self.targetName].values, encoded_feature)[0][1],
                )
            )
        if self.discardOriginal_col:
            X = X.drop(self.targetName, axis=1)
        return X
```

```python
class KFoldTargetEncoderTest(base.BaseEstimator, base.TransformerMixin):
    def __init__(self, train, colNames, encodedName):

        self.train = train
        self.colNames = colNames
        self.encodedName = encodedName

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        mean = (
            self.train[[self.colNames, self.encodedName]]
            .groupby(self.colNames)
            .mean()
            .reset_index()
        )

        dd = {}
        for index, row in mean.iterrows():
            dd[row[self.colNames]] = row[self.encodedName]
        X[self.encodedName] = X[self.colNames]
        X = X.replace({self.encodedName: dd})
        return X
```

```python
X = df_dummy_a
y = df_dummy_a.worldwide_gross_usd.values

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=46
)
```

```python
used_features_a = [
    "worldwide_gross_usd",
    "aspect_ratio",
    "rating",
    "languages",
    "production_countries",
    "production_companies",
    "genres",
    "directors",
    "writers",
]

for x in range(1, len(used_features_a)):
    targetc = KFoldTargetEncoderTrain(
        used_features_a[x], "worldwide_gross_usd", n_fold=5
    )
    X_train_enc = targetc.fit_transform(X_train)

for x in range(1, len(used_features_a)):
    test_targetc = KFoldTargetEncoderTest(
        X_train_enc, "worldwide_gross_usd", used_features_a[x]
    )
    X_test_enc = test_targetc.transform(X_test)
```

# Normalization / Standardization of Features

```python
# Reference (1): https://stackoverflow.com/a/56971065,
# (2): https://towardsai.net/p/data-science/how-when-and-why-should-you-normalize-standardize-rescale-your-data-3f083def38ff
rt = Pipeline(steps=[("robust", RobustScaler())])
mt = Pipeline(steps=[("minmax", MinMaxScaler())])

ct = ColumnTransformer(
    remainder="passthrough",  # "Don't drop any features not mentioned in the transformer"
    transformers=[
        ("robust", rt, ["budget_est_usd", "votes"]),
        (
            "minmax",
            mt,
            [
                "votescore",
                "metacritic",
                "year",
                "runtime_minutes",
                "genre_count",
                "story_word_count",
                "prod_company_count",
                "country_count",
                "language_count",
                "director_count",
                "directors_freq_points",
                "writer_count",
                "writers_freq_points",
            ],
        ),
    ],
    verbose=True,
    verbose_feature_names_out=False,
)
```

```python
X_train_enc.drop(used_features_a, axis=1, inplace=True)
X_test_enc.drop(used_features_a, axis=1, inplace=True)
```

```python
X_train_scaled = ct.fit_transform(X_train_enc)
X_train_sc = pd.DataFrame(X_train_scaled, columns=list(X_train_enc.columns))

X_test_scaled = ct.transform(X_test_enc)
X_test_sc = pd.DataFrame(X_test_scaled, columns=list(X_test_enc.columns))
```

# Linear Regression

```python
lm_a = LinearRegression()
lm_a.fit(X_train_enc, y_train)

# sorted(metrics.SCORERS.keys())
np.mean(
    cross_val_score(lm_a, X_train_sc, y_train, scoring="neg_mean_absolute_error", cv=10)
)
```

# Lasso

```python
la_alpha = []
la_results = []
for x in range(1, 100):
    la_alpha.append(x / 100)
    la_a = Lasso(alpha=x / 100, random_state=46)
    #     la_b.fit(X_train_enc, y_train)
    la_results.append(
        np.mean(
            cross_val_score(
                la_a, X_train_sc, y_train, scoring="neg_mean_absolute_error", cv=10
            )
        )
    )

print(f"alpha = {(la_results.index(max(la_results)) + 1) * .10} \n{max(la_results)}")

plt.plot(la_alpha, la_results)
```

```python
max_y = max(la_results)  # Find the maximum y value
best_alpha = la_alpha[la_results.index(max_y)]  # Find the x value corresponding to the maximum y value
print(best_alpha, max_y)

la_a = Lasso(alpha=best_alpha,random_state=46)
la_a.fit(X_train_sc, y_train)
```

# K-Nearest (A)

```python
knn_neighbors = []
knn_results = []

for i in range(1, 10):
    knn_neighbors.append(i)
    knn_a = KNeighborsRegressor(n_neighbors=i)
    knn_nmae = np.mean(
        cross_val_score(
            knn_a, X_train, y_train, cv=10, scoring="neg_mean_absolute_error"
        )
    )
    knn_results.append(knn_nmae)

plt.plot(knn_neighbors, knn_results)
```

```python
max_y = max(knn_results)
best_i = knn_neighbors[knn_results.index(max_y)]
print(best_i, max_y)

knn_a = KNeighborsRegressor(n_neighbors=best_i)
knn_a.fit(X_train_sc, y_train)
```

# XGBoost (A)

```python
import xgboost as xgb

parameters = {
    "n_estimators": range(180, 201, 10),
    "max_depth": range(3, 10, 2),
    "subsample": np.arange(0.25, 1, 0.2),
    "lambda": (0.1, 0.25, 0.5, 0.75, 1),
    "alpha": (0.1, 0.25, 0.5, 0.75, 1),
}

xgb_a = xgb.XGBRegressor(random_state=46)
gs_xgb_a = RandomizedSearchCV(
    xgb_a,
    parameters,
    n_iter=100,
    scoring="neg_mean_absolute_error",
    cv=10,
    random_state=46,
)
gs_xgb_a.fit(X_train_sc, y_train)
```

```python
from sklearn.ensemble import StackingRegressor

stk_a = StackingRegressor(
    estimators=[
        ("linreg", lm_a),
        ("lasso", la_a),
        ("knn", knn_a),
        ("gs_xgb_a", xgb_best),
        ("gs_gb_a", gb_best),
        ("gs_ada_a", ada_best),
    ],
    final_estimator=GradientBoostingRegressor(random_state=64),
    verbose=False,
)

stk_a.fit(X_train_sc, y_train)
```

# Putting the most efficient model into production

Now to get the most efficient model from each section and saving it using a pickle. This allows us to save time from training models over and over again.

```python
import pickle
def find_efficient_model(criteria, target=1):
    if target == 1:
        if (
            "Gradient Boost"
            == models_a[models_a[criteria] == models_a[criteria].min()]["Model"].iloc[0]
        ):
            best_model = gb_best
        elif (
            "XGBoost"
            == models_a[models_a[criteria] == models_a[criteria].min()]["Model"].iloc[0]
        ):
            best_model = xgb_best

        elif (
            "AdaBoost"
            == models_a[models_a[criteria] == models_a[criteria].min()]["Model"].iloc[0]
        ):
            best_model = ada_best

        elif (
            "KNN"
            == models_a[models_a[criteria] == models_a[criteria].min()]["Model"].iloc[0]
        ):
            best_model = knn_a
        elif (
            "Voting Ensemble"
            == models_a[models_a[criteria] == models_a[criteria].min()]["Model"].iloc[0]
        ):
            best_model = vr_a
        elif (
            "Stack Ensemble"
            == models_a[models_a[criteria] == models_a[criteria].min()]["Model"].iloc[0]
        ):
            best_model = stk_a
        else:
            best_model = lm_a
    else:
        best_model = vr_a

    return best_model


best = find_efficient_model("MAE", 1)

try:
    pickl = {"model": best.best_estimator_}
except AttributeError:
    pickl = {"model": best}

pickle.dump(pickl, open("FlaskAPI/models/" + "model_file_gross" + ".p", "wb"))

file_name = "FlaskAPI/models/model_file_gross.p"
with open(file_name, "rb") as pickled:
    data = pickle.load(pickled)
    model = data["model"]
```
