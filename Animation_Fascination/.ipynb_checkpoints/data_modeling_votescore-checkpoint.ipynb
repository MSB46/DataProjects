# Test Train Split (A)

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

X = df_dummy_a
y = df_dummy_a.worldwide_gross_usd.values

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=46)
```

# Target Encoding (A)

A lot of features have noticeably high cardinality as shown in the EDA section of the project. Therefore, target encoding seems more suitable to me than dummy encoding these features.

### KFoldTargetEncoderTrain by *Anurag Bantu*
#### Source: https://www.kaggle.com/code/anuragbantu/target-encoding-beginner-s-guide

```python
# Credit: anuragbantu
# Source: https://www.kaggle.com/code/anuragbantu/target-encoding-beginner-s-guide

from sklearn import base
from sklearn.model_selection import KFold

class KFoldTargetEncoderTrain(base.BaseEstimator, base.TransformerMixin):
    def __init__(self,colnames,targetName,
                  n_fold=5, verbosity=False,
                  discardOriginal_col=False):
        self.colnames = colnames
        self.targetName = targetName
        self.n_fold = n_fold
        self.verbosity = verbosity
        self.discardOriginal_col = discardOriginal_col
        
    def fit(self, X, y=None):
        return self
    
    def transform(self,X):
        mean_of_target = X[self.targetName].mean()
        kf = KFold(n_splits = self.n_fold, shuffle = True, random_state=46)
        col_mean_name = self.colnames + '_' + 'kfold_target_enc'
        X[col_mean_name] = np.nan
        
        for tr_ind, val_ind in kf.split(X):
            X_tr, X_val = X.iloc[tr_ind], X.iloc[val_ind]
            X.loc[X.index[val_ind], col_mean_name] = X_val[self.colnames].map(X_tr.groupby(self.colnames)
                                     [self.targetName].mean())
            X[col_mean_name].fillna(mean_of_target, inplace = True)
            
        if self.verbosity:
            encoded_feature = X[col_mean_name].values
            print('Correlation between the new feature, {} and, {} is {}.'.format(col_mean_name,self.targetName,                    
                   np.corrcoef(X[self.targetName].values,
                               encoded_feature)[0][1]))
            
        if self.discardOriginal_col:
            X = X.drop(self.targetName, axis=1)
        return X
```

```python
used_features_a = ['worldwide_gross_usd','aspect_ratio', 'rating', 'languages', 'production_countries',
                   'production_companies', 'genres', 'directors', 'writers']

for x in range(1,len(used_features_a)):
    targetc = KFoldTargetEncoderTrain(used_features_a[x],'worldwide_gross_usd',n_fold=5)
    X_train = targetc.fit_transform(X_train)
    X_test = targetc.transform(X_test)
    
    print("------------------")
```

```python
list(X_train.columns)
```

```python
X_train.drop(used_features_a,axis=1,inplace=True)
X_test.drop(used_features_a,axis=1,inplace=True)
```

When using the worldwide gross as the target variable, I get an R-squared value of about 0.697. For adjusted R-squared, I get a value of .659 which seems decent given the high amount of features I've added to the dataframe. I decided to run a few different models to see which one performs best.

* K-Nearest
* Linear Regression (Baseline)
* Random Forest (GridSearchCV)
* Gradient Boost (GridSearchCV)
* Ensemble Methods: Stacking and Voting (KNN + RF + GB)

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```python
from sklearn.linear_model import LinearRegression, Lasso

lm_a = LinearRegression()
lm_a.fit(X_train,y_train)

# sorted(metrics.SCORERS.keys())
np.mean(cross_val_score(lm_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=3))
```

# K-Nearest (A)

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.preprocessing import StandardScaler

knn_neighbors = []
knn_results = []

for i in range(1,10):
    knn_neighbors.append(i)
    knn_a = make_pipeline(StandardScaler(), KNeighborsRegressor(n_neighbors=i))
    knn_nmae = np.mean(cross_val_score(knn_a, X_train, y_train, cv=3, scoring ='neg_mean_absolute_error'))
    knn_results.append(knn_nmae)
    
plt.plot(knn_neighbors,knn_results)
```

```python
max_y = max(knn_results)  
best_i = knn_neighbors[knn_results.index(max_y)]
print(best_i, max_y)

knn_a = make_pipeline(StandardScaler(), KNeighborsRegressor(n_neighbors=best_i))
knn_a.fit(X_train, y_train)
```

# Random Forest (A)

```python
from sklearn.ensemble import RandomForestRegressor

rf_a = RandomForestRegressor(random_state=46)
rf_a.fit(X_train, y_train)

np.mean(cross_val_score(rf_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=3))
```

### Grid Search RF (A)

```python
parameters = {'n_estimators':[80,90,100], 'max_depth':range(3,10,3)}

gs_rf_a = GridSearchCV(rf_a, parameters, scoring='neg_mean_absolute_error', cv=3)
gs_rf_a.fit(X_train,y_train)
```

```python
gs_rf_a.best_score_
```

```python
gs_rf_a.best_estimator_
```

# Gradient Boost (A)

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_a = GradientBoostingRegressor(random_state=46)
gb_a.fit(X_train, y_train)

np.mean(cross_val_score(gb_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=3))
```

### Grid Search CV - GB (A)

```python
parameters = {'n_estimators':[170,180,190],
              'max_depth':[7,9],
              'subsample':np.arange(0.65, 1, 0.15),
              'max_features':['auto'],
              'alpha':(0.1,0.25)}

gs_gb_a = GridSearchCV(gb_a, parameters, scoring='neg_mean_absolute_error', cv=3)
gs_gb_a.fit(X_train,y_train)
```

```python
gs_gb_a.best_score_
```

```python
gs_gb_a.best_estimator_
```

# Voting Regressor (A)

```python
from sklearn.ensemble import VotingRegressor
vr_a = VotingRegressor(estimators = [('knn', knn_a),('gs_rf_a', gs_rf_a.best_estimator_),
                                  ('rs_gb_a',gs_gb_a.best_estimator_)],
                     weights=[1,5,4])
vr_a.fit(X_train,y_train)
```

# Stacking Regressor (A)

```python
from sklearn.ensemble import StackingRegressor

stk_a = StackingRegressor(estimators = [('knn', knn_a), 
                                     ('gs_rf_a', gs_rf_a.best_estimator_),
                                     ('rs_gb_a', gs_gb_a.best_estimator_)],
                          final_estimator = GradientBoostingRegressor(random_state=46),
                          verbose=False)

stk_a.fit(X_train,y_train)
```

# Ensemble Testing (A)

```python
pred_knn = knn_a.predict(X_test)
pred_lm_a = lm_a.predict(X_test)
pred_gs_rf = gs_rf_a.best_estimator_.predict(X_test)
pred_gs_gb = gs_gb_a.best_estimator_.predict(X_test)

pred_vr = vr_a.predict(X_test)
pred_stk = stk_a.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error

print("----MAE----")
print(f'Linear Reg:\t\t{mean_absolute_error(y_test, pred_lm_a)}')
print(f'KNN:\t\t\t{mean_absolute_error(y_test, pred_knn)}')
print(f'Random Forest GS\t{mean_absolute_error(y_test, pred_gs_rf)}')
print(f'Gradient Boosting GS\t{mean_absolute_error(y_test, pred_gs_gb)}')

print(f'VR (KNN+RFGS+GB)\t{mean_absolute_error(y_test, pred_vr)}')
print(f'STK (KNN+RFGS+GB)\t{mean_absolute_error(y_test, pred_stk)}')

print("----RSME----")
print(f'Linear Reg:\t\t{mean_squared_error(y_test, pred_lm_a, squared=False)}')
print(f'KNN:\t\t\t{mean_squared_error(y_test, pred_knn, squared=False)}')
print(f'Random Forest GS:\t{mean_squared_error(y_test, pred_gs_rf, squared=False)}')
print(f'Gradient Boosting GS:\t{mean_squared_error(y_test, pred_gs_gb, squared=False)}')

print(f'VR (KNN+RFGS+GB)\t{mean_squared_error(y_test, pred_vr, squared=False)}')
print(f'STK (KNN+RFGS+GB)\t{mean_squared_error(y_test, pred_stk, squared=False)}')
```

```python
models_a = pd.DataFrame({
    'Model': ['Linear Regression','KNN','Random Forest GS', 'Gradient Boost GS', 'Voting Ensemble','Stack Ensemble'],
    'MAE': [ 
        mean_absolute_error(y_test, pred_lm_a), 
        mean_absolute_error(y_test, pred_knn), 
        mean_absolute_error(y_test, pred_gs_rf), 
        mean_absolute_error(y_test, pred_gs_gb), 
        mean_absolute_error(y_test, pred_vr), 
        mean_absolute_error(y_test, pred_stk), 
    ],    
    'RMSE': [ 
        mean_squared_error(y_test, pred_lm_a, squared=False), 
        mean_squared_error(y_test, pred_knn, squared=False), 
        mean_squared_error(y_test, pred_gs_rf, squared=False), 
        mean_squared_error(y_test, pred_gs_gb, squared=False), 
        mean_squared_error(y_test, pred_vr, squared=False), 
        mean_squared_error(y_test, pred_stk, squared=False), 
    ]
    })
models_a.sort_values(by='MAE', ascending=True)
```

The Voting Ensemble is victorious with both the lowest MAE and RMSE.

```python
la_alpha = []
la_results = []
for x in range(1, 100):
    la_alpha.append(x/100)
    la_b = Lasso(alpha=x/100,random_state=46)
    la_b.fit(X_train, y_train)
    
    la_results.append(np.mean(cross_val_score(la_b, X_train, y_train, scoring='neg_mean_absolute_error', cv=3)))
    
print(f'alpha = {(la_results.index(max(la_results)) + 1) * .10} \n{max(la_results)}')

plt.plot(la_alpha,la_results)
```

```python
max_y = max(la_results)  # Find the maximum y value
best_alpha = la_alpha[la_results.index(max_y)]  # Find the x value corresponding to the maximum y value
print(best_alpha, max_y)

la_b = Lasso(alpha=best_alpha,random_state=46)
la_b.fit(X_train, y_train)
```
