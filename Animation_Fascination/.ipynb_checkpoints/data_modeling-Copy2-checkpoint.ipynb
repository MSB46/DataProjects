# Target Encoding (A)

A lot of features have noticeably high cardinality as shown in the EDA section of the project. Therefore, target encoding seems more suitable to me than dummy encoding these features.

### KFoldTargetEncoderTrain by *Anurag Bantu*
#### Source: https://www.kaggle.com/code/anuragbantu/target-encoding-beginner-s-guide

```python
# Credit: anuragbantu
# Source: https://www.kaggle.com/code/anuragbantu/target-encoding-beginner-s-guide

from sklearn import base
from sklearn.model_selection import KFold

class KFoldTargetEncoderTrain(base.BaseEstimator, base.TransformerMixin):
    def __init__(self,colnames,targetName,
                  n_fold=5, verbosity=True,
                  discardOriginal_col=False):
        self.colnames = colnames
        self.targetName = targetName
        self.n_fold = n_fold
        self.verbosity = verbosity
        self.discardOriginal_col = discardOriginal_col
        
    def fit(self, X, y=None):
        return self
    
    def transform(self,X):
        mean_of_target = X[self.targetName].mean()
        kf = KFold(n_splits = self.n_fold, shuffle = True, random_state=46)
        col_mean_name = self.colnames + '_' + 'kfold_target_enc'
        X[col_mean_name] = np.nan
        
        for tr_ind, val_ind in kf.split(X):
            X_tr, X_val = X.iloc[tr_ind], X.iloc[val_ind]
            X.loc[X.index[val_ind], col_mean_name] = X_val[self.colnames].map(X_tr.groupby(self.colnames)
                                     [self.targetName].mean())
            X[col_mean_name].fillna(mean_of_target, inplace = True)
            
        if self.verbosity:
            encoded_feature = X[col_mean_name].values
            print('Correlation between the new feature, {} and, {} is {}.'.format(col_mean_name,self.targetName,                    
                   np.corrcoef(X[self.targetName].values,
                               encoded_feature)[0][1]))
            
        if self.discardOriginal_col:
            X = X.drop(self.targetName, axis=1)
        return X
```

```python
used_features_a = ['worldwide_gross_usd','aspect_ratio', 'rating', 'languages', 'production_countries',
                   'production_companies', 'genres', 'directors', 'writers']

for x in range(1,len(used_features_a)):
    targetc = KFoldTargetEncoderTrain(used_features_a[x],'worldwide_gross_usd',n_fold=5)
    X_train = targetc.fit_transform(X_train)
    X_test = targetc.fit_transform(X_test)
    
    print("------------------")
```

```python
X_train.drop(used_features_a,axis=1,inplace=True)
X_test.drop(used_features_a,axis=1,inplace=True)
```

```python
X = df_dummy_a.drop(used_features_a,axis=1)

X_sm = sm.add_constant(X)
model = sm.OLS(y, X_sm).fit()
print(model.summary())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```python
from sklearn.linear_model import LinearRegression, Lasso

lm_a = LinearRegression()
lm_a.fit(X_train,y_train)

# sorted(metrics.SCORERS.keys())
np.mean(cross_val_score(lm_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=10))
```

# K-Nearest (A)

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import make_pipeline, Pipeline
from sklearn.preprocessing import StandardScaler

knn_neighbors = []
knn_results = []

for i in range(1,10):
    knn_neighbors.append(i)
    knn_a = make_pipeline(StandardScaler(), KNeighborsRegressor(n_neighbors=i))
    knn_nmae = np.mean(cross_val_score(knn_a, X_train, y_train, cv=10, scoring ='neg_mean_absolute_error'))
    knn_results.append(knn_nmae)
    
plt.plot(knn_neighbors,knn_results)
```

```python
max_y = max(knn_results)  
best_i = knn_neighbors[knn_results.index(max_y)]
print(best_i, max_y)

knn_a = make_pipeline(StandardScaler(), KNeighborsRegressor(n_neighbors=best_i))
knn_a.fit(X_train, y_train)
```

# Random Forest (A)

```python
from sklearn.ensemble import RandomForestRegressor

rf_a = RandomForestRegressor(random_state=46)
rf_a.fit(X_train, y_train)

np.mean(cross_val_score(rf_a, X_train, y_train, scoring='neg_mean_absolute_error', cv=10))
```

### Grid Search RF (A)

```python
parameters = {'n_estimators':[80,90,100], 'max_depth':range(3,10,3)}

gs_rf_a = GridSearchCV(rf_a, parameters, scoring='neg_mean_absolute_error', cv=10)
gs_rf_a.fit(X_train,y_train)
```

# Voting Regressor (A)

```python
models_a = pd.DataFrame({
    'Model': ['Linear Regression','KNN','Random Forest GS', 'Gradient Boost GS', 'Voting Ensemble','Stack Ensemble'],
    'MAE': [ 
        mean_absolute_error(y_test, pred_lm_a), 
        mean_absolute_error(y_test, pred_knn), 
        mean_absolute_error(y_test, pred_gs_rf), 
        mean_absolute_error(y_test, pred_gs_gb), 
        mean_absolute_error(y_test, pred_vr), 
        mean_absolute_error(y_test, pred_stk), 
    ],    
    'RMSE': [ 
        mean_squared_error(y_test, pred_lm_a, squared=False), 
        mean_squared_error(y_test, pred_knn, squared=False), 
        mean_squared_error(y_test, pred_gs_rf, squared=False), 
        mean_squared_error(y_test, pred_gs_gb, squared=False), 
        mean_squared_error(y_test, pred_vr, squared=False), 
        mean_squared_error(y_test, pred_stk, squared=False), 
    ]
    })
models_a.sort_values(by='MAE', ascending=True)
```

The Voting Ensemble is victorious with both the lowest MAE and RMSE.

# Putting the most efficient model into production

Now to get the most efficient model from each section and saving it using a pickle. This allows us to save time from training models over and over again.

```python
import pickle

def find_efficient_model(criteria,target=1):
    if target == 1:
        if "Gradient Boost GS" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = gs_gb_b
        elif "CatBoost GS" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = gs_cbr_b
        elif "KNN" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = knn_b
        elif "Voting Ensemble" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = vr_b
        elif "Stack Ensemble" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = stk_b
        elif "Linear Regression" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = lm_b
        elif "Lasso" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = la_b
        else:
            best_model = gs_xgb_b
    else:
        best_model = vr_a
        
    return best_model

best = find_efficient_model("MAE",1)
    
try:
    pickl = {'model': best.best_estimator_}
except AttributeError:
    pickl = {'model': best}
    
pickle.dump(pickl, open('FlaskAPI/models/'+'model_file' + ".p", "wb"))

file_name = "FlaskAPI/models/model_file.p"
with open(file_name, 'rb') as pickled:
    data = pickle.load(pickled)
    model = data['model']
```
