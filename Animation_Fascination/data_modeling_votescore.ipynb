### KFoldTargetEncoderTrain by *Anurag Bantu*
#### Source: https://www.kaggle.com/code/anuragbantu/target-encoding-beginner-s-guide

```python
# Credit: anuragbantu
# Source: https://www.kaggle.com/code/anuragbantu/target-encoding-beginner-s-guide

from sklearn import base
from sklearn.model_selection import KFold

class KFoldTargetEncoderTrain(base.BaseEstimator, base.TransformerMixin):
    def __init__(self,colnames,targetName,
                  n_fold=5, verbosity=False,
                  discardOriginal_col=False):
        self.colnames = colnames
        self.targetName = targetName
        self.n_fold = n_fold
        self.verbosity = verbosity
        self.discardOriginal_col = discardOriginal_col
        
    def fit(self, X, y=None):
        return self
    
    def transform(self,X):
        mean_of_target = X[self.targetName].mean()
        kf = KFold(n_splits = self.n_fold, shuffle = True, random_state=46)
        col_mean_name = self.colnames + '_' + 'kfold_target_enc'
        X[col_mean_name] = np.nan
        
        for tr_ind, val_ind in kf.split(X):
            X_tr, X_val = X.iloc[tr_ind], X.iloc[val_ind]
            X.loc[X.index[val_ind], col_mean_name] = X_val[self.colnames].map(X_tr.groupby(self.colnames)
                                     [self.targetName].mean())
            X[col_mean_name].fillna(mean_of_target, inplace = True)
            
        if self.verbosity:
            encoded_feature = X[col_mean_name].values
            print('Correlation between the new feature, {} and, {} is {}.'.format(col_mean_name,self.targetName,                    
                   np.corrcoef(X[self.targetName].values,
                               encoded_feature)[0][1]))
            
        if self.discardOriginal_col:
            X = X.drop(self.targetName, axis=1)
        return X
```

```python
class KFoldTargetEncoderTest(base.BaseEstimator, base.TransformerMixin):
    def __init__(self, train, colNames, encodedName):

        self.train = train
        self.colNames = colNames
        self.encodedName = encodedName

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        mean = (
            self.train[[self.colNames, self.encodedName]]
            .groupby(self.colNames)
            .mean()
            .reset_index()
        )

        dd = {}
        for index, row in mean.iterrows():
            dd[row[self.colNames]] = row[self.encodedName]
        X[self.encodedName] = X[self.colNames]
        X = X.replace({self.encodedName: dd})
        return X
```

```python
# X = df_dummy_b.drop(['votescore'],axis=1)
# y = df_dummy_b.votescore.values

X = df_dummy_b
y = df_dummy_b.votescore.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=46)
```

# Target Encoding (B)

```python
used_features_b = ['votescore','aspect_ratio', 'rating', 'languages', 'production_countries',
                   'production_companies', 'genres', 'directors','writers']

for x in range(1,len(used_features_b)):
    targetc = KFoldTargetEncoderTrain(used_features_b[x],'votescore',n_fold=5)
    X_train_enc = targetc.fit_transform(X_train)
#     X_test = targetc.transform(X_test)
    
for x in range(1, len(used_features_b)):
    test_targetc = KFoldTargetEncoderTest(
        X_train_enc, "votescore", used_features_b[x] + "_kfold_target_enc"
    )
    X_test_enc = test_targetc.transform(X_test)
```

```python
X_train_enc.drop(used_features_b,axis=1,inplace=True)
X_test_enc.drop(used_features_b,axis=1,inplace=True)
```

```python
X_train_final = X_train_enc
X_test_final = X_test_enc
```

# Putting the most efficient model into production

Now to get the most efficient model from each section and saving it using a pickle. This allows us to save time from training models over and over again.

```python
import pickle

def find_efficient_model(criteria,target=1):
    if target == 1:
        if "Gradient Boost GS" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = gb_best
        elif "CatBoost GS" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = cb_best
        elif "KNN" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = knn_best
        elif "Voting Ensemble" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = vr_b
        elif "Stack Ensemble" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = stk_b
        elif "Linear Regression" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = lm_b
        elif "Lasso" == models_b[models_b[criteria] == models_b[criteria].min()]['Model'].iloc[0]:
            best_model = la_best
        else:
            best_model = xgb_best
    else:
        best_model = vr_a
        
    return best_model

best = find_efficient_model("MAE",1)
    
try:
    pickl = {'model': best.best_estimator_}
except AttributeError:
    pickl = {'model': best}
    
pickle.dump(pickl, open('FlaskAPI/models/'+'model_file_votescore' + ".p", "wb"))

file_name = "FlaskAPI/models/model_file_votescore.p"
with open(file_name, 'rb') as pickled:
    data = pickle.load(pickled)
    model = data['model']
```
